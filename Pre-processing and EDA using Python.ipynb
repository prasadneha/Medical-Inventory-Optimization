# Removing Outliers

***The Z-score represents the number of standard deviations a data point is 
from the mean. Typically, a Z-score above 3 or below -3 is 
considered an outlier.***


```python
# Displaying the shape of the cleaned_data DataFrame to 
# understand its dimensions before removing outliers
cleaned_data.shape
```

```python
from scipy import stats

# Calculating z-scores for numerical columns in the cleaned_data DataFrame
# Z-scores are a measure of how many standard deviations away a data point is from the mea
z_scores = np.abs(stats.zscore(cleaned_data.select_dtypes(include=[np.number])))

# Identify outliers (threshold: z-score > 3)
# This is a common threshold for defining outliers in a dataset
outliers = (z_scores > 3)
print(outliers.head(20))

# Remove rows with any outliers from the cleaned_data DataFrame
# The ~ operator negates the boolean values, and .any(axis=1) checks for any True values across each row
Final_data = cleaned_data[~outliers.any(axis=1)]

# Display the first 20 rows of the resulting DataFrame
# print(Final_data.head(20))
```
